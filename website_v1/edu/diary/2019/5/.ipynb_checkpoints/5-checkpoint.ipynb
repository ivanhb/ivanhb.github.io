## DIPAM main new features

* Now DIPAM will save also the layout and elements position. Therefor, when openning (Loading) a new workflow the diagram elements will appear in the same position as they were when saving the workflow.  
* It is possible to specify the parameters to use for the corresponding tools.The parameters should have a “type” attribute also specified in the configuration file, along with the type of HTML dom used to specifiy/define such parameter (e.g. dropdown list). ## DIPAM for CATARSI: Server side
When generating the files of each different tool in its corresponding directory, these files will be named following a specific pattern (file name) specified in the configuration file.

## DIAPM Demo 
The current Demo can perform a LDA Topic Modelling algorithm and Filtering options on a Textual file and make it available to download. Along with defining some parameters for each different tool used.  
To make this we have defined in the configuration file "Data" as:
```
        "d-topics-table":{
            "label": "Table of topics",
            "data_class": "table",
            "param": ["p-file"],
            "file_name": "topics"
        },
        "d-gen-text":{
            "label": "Text file",
            "data_class": "text",
            "param": ["p-file"],
            "file_name": "doc"
        },
        "d-coherence":{
            "label": "Topic coherence value",
            "data_class": "text",
            "param": ["p-file"],
            "file_name": "coherence"
        }
```
And the tools as:
```
        "t-topic-lda": {
            "label": "Topic modeling with LDA",
            "function": "lda",
            "param": ["p-topic","p-numwords"],
            "compatible_input": ["d-gen-text"],
            "output": ["d-topics-table","d-coherence"]
        },
        "t-chart-bar": {
            "label": "Bar chart",
            "function": "bar_chart",
            "param": ["p-log"],
            "compatible_input": ["d-topics-table"]
        },
        "t-filter-text": {
            "label": "Filter the text",
            "function": "filter_text",
            "param": ["p-filteropt","p-filterregex"],
            "compatible_input": ["d-gen-text"],
            "output": ["d-gen-text"]
        },
        "t-save-files": {
            "label": "Save",
            "function": "save_file",
            "compatible_input": ["d-gen-text","d-topics-table","d-coherence"]
        }
```
Along with the all the new parameters used:
```
        "p-log":{
          "handler": "switch-value",
          "label": "Log scale",
          "value_label": ["True","False"],
          "value": [0,1],
          "init_value": 0
        },
        "p-topic":{
          "handler": "select-value",
          "label": "Number of topics",
          "value_label": ["2","3","4","5","10","15","20"],
          "value": [2,3,4,5,10,15,20],
          "init_value": 2
        },
        "p-numwords":{
          "handler": "select-value",
          "label": "Number of words",
          "value_label": ["10","15","20","25","30","40","50"],
          "value": [10,15,20,25,30,40,50],
          "init_value": 10
        },
        "p-filteropt":{
          "handler": "check-value",
          "label": "Filter and remove:",
          "value_label": ["Dates","Footnotes"],
          "value": ["dates","footnotes"],
          "init_value": ["dates"]
        },
        "p-filterregex":{
          "handler": "input-text",
          "label": "Specific text <p> (Regular expressions supported) </p>",
          "value": "",
          "init_value": ""
        }
```
