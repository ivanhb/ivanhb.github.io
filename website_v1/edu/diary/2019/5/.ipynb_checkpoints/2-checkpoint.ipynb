## DIPAM for CATARSI

Some of the main new additions made to DIPAM are:
* Integrating Flask: The ’Save’ and ‘Load’ operations are now available. 
* The ’Save’ button should open a file dialog window to let users decide in which directory the corresponding file must be saved.  
* No more “Paths” to order the execution. Now I use only the nodes ordered in topological way. 
* For each different node in the topological processing order of the nodes the server will be called with a POST call: /process . When calling this on the client side, the node data should be submitted. The server need to keep track also for the input data needed for such node to be correctly executed.
* On the server side we have two basic directories: .data/ and .process-temp/ . The first directory is used to hold the current workflow.json and config.json, used in DIPAM. When a user want to LOAD a new workflow the inner workflow file in this directory will be subscribed. The second directory is used to save the data (outputs) generated by each node of the process. Each node will have a different directory named according to its ID, these directory will contain the results obtained after the execution of such node. 

### Notes:
* ‘Load’ and ’Save’ workflow now work. But the ’Save’ button still need to be fixed when choosing the saving directory. The problem here is due to the fact that we can't send the directory path chosen on client side to the server side, since its a limitation as a consequence of privacy limitation. A possible solution is to save the workflow in a fixed directory.
* Many problems have raised when integrating flask and making DIPAM relate on the JSON files uploaded. Therefor, important adjustments have been made to the general code logic and structure. This part to a lot of time, and is still a work in progress.
